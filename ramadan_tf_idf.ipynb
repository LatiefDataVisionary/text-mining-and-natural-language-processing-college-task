```python
import pandas as pd
import numpy as np

dm = pd.read_excel("/content/ramadan_clean_tweet_label.xlsx", usecols=["tweet_clean", "Label"])
dm.columns = ["tweet_clean", "Label"]
dm.head(10)
```

```python
# Convert string jadi lsit
import ast

def convert_text_list(texts):
  texts = ast.literal_eval(texts)
  return [text for text in texts]

dm["tekslist"] = dm["tweet_clean"].apply(convert_text_list)

print(dm['tekslist'][23])

print("\ntype: ", type(dm["tekslist"][23]))
```

```python
def calc_TF(document):
  # Perhitungan jumlah kata
  TF_dict = {}
  for term in document:
    if term in TF_dict:
      TF_dict[term] += 1
    else:
      TF_dict[term] = 1
  # Perhitungan tf
  for term in TF_dict:
    TF_dict[term] = TF_dict[term] / len(document)
  return TF_dict

dm["TF_dict"] = dm["tekslist"].apply(calc_TF)

dm["TF_dict"].head()
```

```python
index = 23

print("%20s" % "term", "\t", "TF\n")
for key in dm["TF_dict"][index]:
  print("%20s" % key, "\t", dm["TF_dict"][index][key])
```

```python
def calc_DF(tfDict):
  count_DF = {}
  for document in tfDict:
    print(f"doc : {document}")
    for term in document:
      print(f"term : {term}")
      if term in count_DF:
        count_DF[term] += 1
      else:
        count_DF[term] = 1
  return count_DF


DF = calc_DF(dm["TF_dict"])
```

```python
# Menghitung idf

n_document = len(dm)
def calc_IDF(__n_document, __DF):
  IDF_Dict = {}
  for term in __DF:
    IDF_Dict[term] = np.log(__n_document / (__DF[term] + 1))
  return IDF_Dict

# Penyimpanan kamus idf
IDF = calc_IDF(n_document, DF)
```

```python
import pandas as pd
import numpy as np

dm = pd.read_csv("/content/ramadan_labeled_sentiment.csv", usecols=["tweet_clean", "sentiment"])
dm.columns = ["tweet_clean", "sentiment"]
dm.head(10)
```

```python
import ast

def join_text_list(texts):
    texts = ast.literal_eval(texts)
    return ' '.join([text for text in texts])
dm["tweet_join"] = dm["tweet_clean"].apply(join_text_list)

dm["tweet_join"].head()
```

Menghitung TF-IDF menggunakan TfidfVectorizer
Untuk menghitung TF-IDF menggunakan Scikit-Learn, dapat dilakukan dengan cara berikut,

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# banyaknya term yang akan digunakan,
# di pilih berdasarkan top max_features
# yang diurutkan berdasarkan term frequency seluruh corpus
max_features = 10000

# Feature Engineering
print ("------- TF-IDF on Tweet data -------")

tf_idf = TfidfVectorizer(max_features=max_features, binary=True)
tf_idf.fit(dm["tweet_join"])  # Melakukan fitting terlebih dahulu
tfidf_mat = tf_idf.transform(dm["tweet_join"]).toarray()

print("TF-IDF ", type(tfidf_mat), tfidf_mat.shape)
```

Hasil dari tfidf menggunakan dataframe

```python
import pandas as pd

# Mengambil nama fitur (term)
feature_names = tf_idf.get_feature_names_out()

# Membuat DataFrame
df_tfidf = pd.DataFrame(tfidf_mat, columns=feature_names)

# Menampilkan DataFrame
display(df_tfidf.head(20))
```

melihat list top 50 term yang meiliki TF-IDF terbesar

```python
terms = tf_idf.get_feature_names_out()

# sum tfidf frequency of each term through documents
sums = tfidf_mat.sum(axis=0)

# connecting term to its sums frequency
data = []
for col, term in enumerate(terms):
    data.append((term, sums[col] ))

ranking = pd.DataFrame(data, columns=['term','rank'])
ranking.sort_values('rank', ascending=False)
```

```python
import pandas as pd
from transformers import pipeline

# Load your CSV data
df = dm

# Initialize the pretrained sentiment analysis pipeline
# This pipeline uses a model already trained for sentiment classification.
classifier = pipeline("sentiment-analysis")

# Function to join tokens and get sentiment prediction
def label_tweet(tokens):
    # If tokens are stored as a string representation of a list, you may need to eval() it.
    if isinstance(tokens, str):
        # Use eval with caution; ensure your data is trusted.
        tokens = eval(tokens)
    text = " ".join(tokens)
    result = classifier(text)
    return result[0]['label']  # returns 'POSITIVE' or 'NEGATIVE'

# Apply the function to the 'tweet_clean' column
df['predicted_label'] = df['tweet_clean'].apply(label_tweet)

# Save the results to a new CSV file
df.to_csv("labeled_tweets.csv", index=False)

print("Labeling complete. Check the 'labeled_tweets.csv' file for results.")
```
